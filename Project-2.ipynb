```python
import pandas as pd
import random

# Create a dictionary with sample data
data = {
    'Registration Number': [f'{i+1:02d}' for i in range(50)],
    'Name': [f'Student_{i+1}' for i in range(50)],
    'TT-1 Marks':[random.randint(10, 20) for i in range(50)] 
}

# Convert the dictionary to a DataFrame
df_tt1 = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df_tt1.to_csv('term-test-1-result.csv', index=False)
df_tt1
```

```python
# Create a new DataFrame with 'Registration Number' and 'Name' columns
df_tt2 = df_tt1[['Registration Number','Name']]

# Add the 'TT-2 Marks' column with different values
df_tt2['TT-2 Marks'] = [random.randint(10, 20) for i in range(50)]

# Save the new DataFrame to a CSV file
df_tt2.to_csv('term-test-2-result.csv', index=False)
df_tt2
```

```python
# Load both CSV files into different DataFrames
df_tt1 = pd.read_csv('term-test-1-result.csv')
df_tt2 = pd.read_csv('term-test-2-result.csv')

# Merge the DataFrames on the 'Registration Number' column
df_merged = pd.merge(df_tt1, df_tt2, on='Registration Number')

# Display the merged DataFrame
print(df_merged.head())
```

```python
# Create a new column with the best term test marks for each student
df_merged['Best TT Marks'] = df_merged[['TT-1 Marks', 'TT-2 Marks']].max(axis=1)

# Create a new column with the average term test marks for each student
df_merged['Average TT Marks'] = df_merged[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1)

# Display the first few rows of the updated DataFrame
print(df_merged.head())
```

```python
# Drop the 'TT-1 Marks' and 'TT-2 Marks' columns
df_merged.drop(columns=['TT-1 Marks', 'TT-2 Marks'], inplace=True)

# Display the first few rows of the updated DataFrame
print(df_merged.head())
```

```python
# Write the content of the DataFrame to a new CSV file
df_merged.to_csv('final term test result.csv', index=False)
df_merged
```

```python
X = df_merged.drop(columns=['Name_x','Name_y','Best TT Marks'], axis=1).values
X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss=[]
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=2)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('wcss')
plt.show()
```

```python
# Apply K-Means clustering with K=3 (you can change this value to experiment with different numbers of clusters)
kmeans = KMeans(n_clusters=3, random_state=42)
Y = kmeans.fit_predict(X)
Y
```

```python
kmeans.cluster_centers_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='red', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='pink', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='yellow', label='Cluster 3')


plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c='black', label='Centroids')
plt.title('Clusters of Average TT Marks')
plt.xlabel('Average TT Marks')
plt.ylabel('Registration number')
plt.show()
```
